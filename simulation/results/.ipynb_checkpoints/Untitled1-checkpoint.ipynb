```python
from decimal import Decimal

data = {
    'processor_log_mct': {
        'task_name': ['job6', 'job1', 'job5', 'job4', 'job3', 'job2'],
        'task_start_time': [0, 0, Decimal('2.8'), Decimal('1.2'), 0, 0],
        'task_end_time': [Decimal('2.8'), Decimal('1.2'), Decimal('3.8'), Decimal('3.2'), Decimal('1.6'), Decimal('1.9')],
        'processor_name': ['Processor 1', 'Processor 2', 'Processor 1', 'Processor 2', 'Processor 3', 'Processor 4']
    },
    'processor_log_ga': {
        'task_name': ['job1', 'job2', 'job3', 'job4', 'job5', 'job6'],
        'task_start_time': [0, 0, Decimal('1.2'), 0, Decimal('1.9'), 0],
        'task_end_time': [Decimal('1.2'), Decimal('1.9'), Decimal('2.8'), Decimal('2.0'), Decimal('2.9'), Decimal('2.8')],
        'processor_name': ['Processor 3', 'Processor 2', 'Processor 3', 'Processor 4', 'Processor 2', 'Processor 1']
    },
    'processor_log_pso': {
        'task_name': ['job1', 'job2', 'job3', 'job4', 'job5', 'job6'],
        'task_start_time': [0, 0, Decimal('1.2'), 0, Decimal('1.9'), 0],
        'task_end_time': [Decimal('1.2'), Decimal('1.9'), Decimal('2.8'), Decimal('2.0'), Decimal('2.9'), Decimal('2.8')],
        'processor_name': ['Processor 1', 'Processor 3', 'Processor 1', 'Processor 4', 'Processor 3', 'Processor 2']
    },
    'num_tasks': 6
}

algorithms = ['MCT', 'GA', 'PSO']

for algorithm in algorithms:
    processor_logs = data.get(f'processor_log_{algorithm.lower()}')
    if processor_logs is None:
        continue

    task_names = processor_logs['task_name']
    task_start_times = processor_logs['task_start_time']
    task_end_times = processor_logs['task_end_time']
    processor_names = processor_logs['processor_name']

    print(f"Algorithm: {algorithm}")
    print("Processor Logs:")

    waiting_times = []
    response_times = []

    for i in range(len(task_names)):
        
        task_name = task_names[i]
        task_start_time = task_start_times[i]
        task_end_time = task_end_times[i]
        processor_name = processor_names[i]
        
        response_time = task_start_time # The response time can be calculated as the difference between the task start time and 
                                        # the time when the task actually starts executing on the assigned processor
            
        turnaround_time = task_end_time - task_start_time # The turnaround time can be calculated as the difference between 
                                                          # the task's end time and its start time
            
        waiting_time = response_time # The waiting time can be calculated as the difference between the 
                                     # response time and the start time of the task
            
        print(f"Task: {task_name}, Processor: {processor_name}")
        print(f"Start Time: {task_start_time}, End Time: {task_end_time}")
        print(f"Response Time: {response_time}, Turnaround Time: {turnaround_time}")
        print(f"Waiting Time: {waiting_time}\n\n")

        waiting_times.append(waiting_time)
        response_times.append(response_time)

    total_time = max(task_end_times)
    cpu_utilization = min((sum(float(turnaround_time) for turnaround_time in task_end_times) / float(total_time)) * 100, 100)
    throughput = len(task_names) / total_time
    average_turnaround_time = sum(turnaround_time for turnaround_time in task_end_times) / len(task_names)
    average_waiting_time = sum(waiting_time for waiting_time in waiting_times) / len(task_names)
    average_response_time = sum(response_time for response_time in response_times) / len(task_names)
    
    idle_time = float(total_time) - sum(float(turnaround_time) for turnaround_time in task_end_times)
    
    print("Metrics:")
    print(f"CPU Utilization: {cpu_utilization:.2f}%")
    print(f"Throughput: {throughput:.2f} tasks per unit of time")
    print(f"Average Turnaround Time: {average_turnaround_time:.2f}")
    print(f"Average Waiting Time: {average_waiting_time:.2f}")
    print(f"Average Response Time: {average_response_time:.2f}")
    print(f"Idle Time: {idle_time:.2f}\n")

```

```python
from decimal import Decimal
from collections import defaultdict

data = {
    'processor_log_mct': {
        'task_name': ['job6', 'job1', 'job5', 'job4', 'job3', 'job2'],
        'task_start_time': [0, 0, Decimal('2.8'), Decimal('1.2'), 0, 0],
        'task_end_time': [Decimal('2.8'), Decimal('1.2'), Decimal('3.8'), Decimal('3.2'), Decimal('1.6'), Decimal('1.9')],
        'processor_name': ['Processor 1', 'Processor 2', 'Processor 1', 'Processor 2', 'Processor 3', 'Processor 4']
    },
    'processor_log_ga': {
        'task_name': ['job1', 'job2', 'job3', 'job4', 'job5', 'job6'],
        'task_start_time': [0, 0, Decimal('1.2'), 0, Decimal('1.9'), 0],
        'task_end_time': [Decimal('1.2'), Decimal('1.9'), Decimal('2.8'), Decimal('2.0'), Decimal('2.9'), Decimal('2.8')],
        'processor_name': ['Processor 3', 'Processor 2', 'Processor 3', 'Processor 4', 'Processor 2', 'Processor 1']
    },
    'processor_log_pso': {
        'task_name': ['job1', 'job2', 'job3', 'job4', 'job5', 'job6'],
        'task_start_time': [0, 0, Decimal('1.2'), 0, Decimal('1.9'), 0],
        'task_end_time': [Decimal('1.2'), Decimal('1.9'), Decimal('2.8'), Decimal('2.0'), Decimal('2.9'), Decimal('2.8')],
        'processor_name': ['Processor 1', 'Processor 3', 'Processor 1', 'Processor 4', 'Processor 3', 'Processor 2']
    },
    'num_tasks': 6
}

algorithms = ['MCT', 'GA', 'PSO']

for algorithm in algorithms:
    processor_logs = data.get(f'processor_log_{algorithm.lower()}')
    if processor_logs is None:
        continue

    task_names = processor_logs['task_name']
    task_start_times = processor_logs['task_start_time']
    task_end_times = processor_logs['task_end_time']
    processor_names = processor_logs['processor_name']

    print(f"Algorithm: {algorithm}")
    print("Processor Logs:")

    waiting_times = []
    response_times = []
    
    for i in range(len(task_names)):
        task_name = task_names[i]
        task_start_time = task_start_times[i]
        task_end_time = task_end_times[i]
        processor_name = processor_names[i]

        response_time = task_start_time
        turnaround_time = task_end_time - task_start_time
        waiting_time = response_time

        print(f"Task: {task_name}, Processor: {processor_name}")
        print(f"Start Time: {task_start_time}, End Time: {task_end_time}")
        print(f"Response Time: {response_time}, Turnaround Time: {turnaround_time}")
        print(f"Waiting Time: {waiting_time}\n\n")
        
        idle_time = float(task_start_time - task_end_times[i - 1]) if i > 0 else float(task_start_time)
        if idle_time < 0:
            idle_time = 0
            
        waiting_times.append(waiting_time)
        response_times.append(response_time)

    # Calculate idle time for each processor
    processor_idle_times = defaultdict(float)

    for i in range(len(task_names)):
        task_name = task_names[i]
        task_start_time = task_start_times[i]
        task_end_time = task_end_times[i]
        processor_name = processor_names[i]



        processor_idle_times[processor_name] += idle_time

    total_idle_time = sum(processor_idle_times.values())
    total_time = max(task_end_times)
    cpu_utilization = min((sum(float(turnaround_time) for turnaround_time in task_end_times) / float(total_time)) * 100, 100)
    throughput = len(task_names) / total_time
    average_turnaround_time = sum(turnaround_time for turnaround_time in task_end_times) / len(task_names)
    average_waiting_time = sum(waiting_time for waiting_time in waiting_times) / len(task_names)
    average_response_time = sum(response_time for response_time in response_times) / len(task_names)
    
    print("Metrics:")
    print(f"CPU Utilization: {cpu_utilization:.2f}%")
    print(f"Throughput: {throughput:.2f} tasks per unit of time")
    print(f"Average Turnaround Time: {average_turnaround_time:.2f}")
    print(f"Average Waiting Time: {average_waiting_time:.2f}")
    print(f"Average Response Time: {average_response_time:.2f}")
    print(f"Total Idle Time: {total_idle_time:.2f}")

    # Individual processor idle times
    for processor_name, idle_time in processor_idle_times.items():
        print(f"Idle Time for {processor_name}: {idle_time:.2f}")
        
    print("\n\n")
```
